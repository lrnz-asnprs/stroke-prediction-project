```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn_extra.cluster import KMedoids
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn import metrics
from scipy.spatial.distance import cdist
from sklearn.metrics import silhouette_score
from sklearn import datasets
```

```python
df = pd.read_csv("healthcare-dataset-stroke-data-kopi.csv")
```

```python
df = df.fillna(df.median())
```

```python
df.isnull().sum().sum()
```

# Glucose/bmi, kmeans & elbow + silhouette

```python
# Elbow 

gluc_bmi = df[['bmi', 'avg_glucose_level']]
distortions = [] #Distortion: It is calculated as the average of the squared distances from the cluster centers 
                 #of the respective clusters. Typically, the Euclidean distance metric is used.
K = range(1, 10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(gluc_bmi)
    kmeanModel.fit(gluc_bmi)
    distortions.append(kmeanModel.inertia_)

plt.plot(K, distortions, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
```

```python
#silhouette

sildist = []

Kk = range(2, 10)

for k in Kk: 
    kmeanModel = KMeans(n_clusters=k).fit(gluc_bmi)
    #print(silhouette_score(gluc))
    labels = kmeanModel.labels_
    sildist.append(silhouette_score(gluc_bmi, labels, metric='euclidean'))
    
plt.plot(Kk, sildist, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Silhouette Distorion')
plt.title('The Silhouette Method using Distortion')
plt.show()
```

```python
#2 clusters 
kmeans = KMeans(init="random", n_clusters=2, n_init=10, max_iter=300, random_state=42)
kmeans.fit(gluc_bmi)
kmeans_predict = kmeans.predict(gluc_bmi)

plt.scatter(gluc_bmi["bmi"], gluc_bmi["avg_glucose_level"], c=kmeans_predict)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1]);
```

```python
score = silhouette_score(df[['bmi', 'avg_glucose_level']], kmeans.labels_, metric='euclidean')
score
```

```python
#3 clusters 
kmeans = KMeans(init="random", n_clusters=3, n_init=10, max_iter=300, random_state=42)
kmeans.fit(gluc_bmi)
kmeans_predict = kmeans.predict(gluc_bmi)

plt.scatter(gluc_bmi["bmi"], gluc_bmi["avg_glucose_level"], c=kmeans_predict)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1]);
```

Silhouette evaluation

```python
score = silhouette_score(df[['bmi', 'avg_glucose_level']], kmeans.labels_, metric='euclidean')
score
```
